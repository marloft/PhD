```python
import pandas as pd
import numpy as np

train = pd.read_csv('SumStudentClicksxResult-2-0-50.csv')   # better be in the correct directory!
test = pd.read_csv('SumStudentClicksxResult-2-50-100.csv')

train.head()
```

```python
test.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

cols = ['SumOfsum_click']
cols2 = ['final_result']
trainArr = train.as_matrix(cols)
trainRes = train.as_matrix(cols2)

testArr = test.as_matrix(cols)
testRes = test.as_matrix(cols2)

knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
knn.fit(trainArr, trainRes)

output = knn.predict(testArr)

# or predict on a specific example!
for i in range(300):
    if testRes[i][0] == output[i]:
    #if testRes[i][0] == "Distinction":
    #if output[i] == "Distinction":
        print("Prediction: " + str(output[i]) + "   Clicks: " + str(testArr[i]) + "   Actual: " + str(testRes[i]) + "   CORRECT")
    else:
        print("Prediction: " + str(output[i]) + "   Clicks: " + str(testArr[i]) + "   Actual: " + str(testRes[i]))

        
    
    
    
```

```python
correct = 0.0

for i in range(len(output)): 
    if testRes[i][0] == output[i]: 
        correct += 1
    
correct / len(output)
```
